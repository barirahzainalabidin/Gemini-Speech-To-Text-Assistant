```python
import os
import subprocess
import pathlib
import textwrap
import google.generativeai as genai
from IPython.display import display, Markdown, Audio
from gtts import gTTS
import speech_recognition as sr

# Configure Google Gemini API Key
GOOGLE_GEMINI_API_KEY = os.environ['GOOGLE_GEMINI_API_KEY']
genai.configure(api_key=GOOGLE_GEMINI_API_KEY)
model = genai.GenerativeModel('gemini-pro')

# For Colab, use input text instead of using microphone.
while True:
    user_input = input("Type your query to Gemini (or type 'exit' to stop): ")
    if user_input.lower() == 'exit':
        break
    # Generate response from Gemini API
    response = model.generate_content(user_input)

    # Text-to-Speech using gTTS
    speech = gTTS(response.text, lang='en')
    speech.save('output.mp3')

    # Play the generated audio
    display(Audio('output.mp3', autoplay=True))
```
